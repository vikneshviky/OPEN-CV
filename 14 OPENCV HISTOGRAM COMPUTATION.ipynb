```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('photos/cards.jfif')

cv.imshow('people',img)
cv.waitKey(0)
cv.destroyAllWindows
```

```python
#conver img to grayscale
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow('gray',img_gray)
cv.waitKey(0)
cv.destroyAllWindows
```

```python
#grayscale to histogram
gray_hist = cv.calcHist([img_gray],[0],None,[256],[0,256])

plt.figure()
plt.title("gray scale hist")
plt.xlabel('bins')
plt.ylabel('no of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()
```

```python
#mask to histogram
#get the codes from mask
blank = np.zeros(img.shape[:2],dtype='uint8')
circle = cv.circle(blank,(img.shape[1]//2,img.shape[0]//2),100,255,-1)
gray_hist = cv.calcHist([img_gray],[0],circle,[256],[0,256])

mask = cv.bitwise_and(img_gray,img_gray,mask=circle)

plt.figure()
plt.title("gray scale hist")
plt.xlabel('bins')
plt.ylabel('no of pixels')
plt.plot(gray_hist)
plt.xlim([0,256])
plt.show()

cv.imshow('mask',mask)
cv.waitKey(0)
cv.destroyAllWindows
```

```python
#color image histogram
circle_mask = cv.circle(blank,(img.shape[1]//2,img.shape[0]//2),100,255,-1)

plt.figure()
plt.title("Color histogram")
plt.xlabel('bins')
plt.ylabel('no of pixels')

colors = ('b','g','r')
for i,col in enumerate(colors):
    hist = cv.calcHist([img],[i],circle_mask,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])

plt.show()
          
```
