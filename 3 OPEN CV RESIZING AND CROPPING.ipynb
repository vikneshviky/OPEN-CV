# OPEN CV CONVENTION


the above image is 640 x 480. the origin of this image is (0,0). max height and width is (640 x480)


```python
import cv2 as cv
```

```python
img = cv.imread('photos/corv.webp')
print(img.shape)#show the size of the image

img_resize = cv.resize(img,(300,200)) #size is reduced / W x H
print(img_resize.shape)#show the size of resized image

img_crop = img[0:200,200:500] #here H comes first H x W

cv.imshow('Car',img)
cv.imshow('Resize car',img_resize)
cv.imshow('Cropped imaged',img_crop)


cv.waitKey(0)
```
